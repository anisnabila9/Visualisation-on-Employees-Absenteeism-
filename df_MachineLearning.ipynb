## Standardize the data

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        self.copy = copy  # Add this line
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
    
    def get_params(self, deep=True):
        return {'columns': self.columns, 'copy': self.copy, 'with_mean': self.scaler.with_mean, 'with_std': self.scaler.with_std}

    def set_params(self, **params):
        self.columns = params['columns']
        self.copy = params['copy']
        self.scaler.with_mean = params['with_mean']
        self.scaler.with_std = params['with_std']
        return self
```

## Interpreting the coefficients

Standardized coefficients are basically : the coefficients of a regression where all variables have been standardized

```python
# create a new Series called: 'Odds ratio' which will show the.. odds ratio of each feature
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
# sort the table according to odds ratio
# note that by default, the sort_values method sorts values by 'ascending'
summary_table.sort_values('Odds_ratio', ascending=False)
```

Interpretation :

a feature is not particularly important :
- coeff around = 0
- odds ration around = 1 

A weight(coefficient) of 0 implies that no matter the feature value, we will multiply it by 0(in the model)

For a unitt change in the standardize feature, the odds increase by a multiple equal to the odds ratio (1=no change)

#### `Standardize model` often leads to lessen interpretable model(result not very clear) but higher accuracy

#### `Machine Learning Engineers` : prefer models with higher accuracy, so normally go w standardization.

#### `Econometricians and Stastisticians` : prefer less accuracy but more interpretavle models, bcz they care about underlying reasons behind different phenomenoa.

#### `Data Scientist` : either position. Sometimes they need higher accuracy, other times- must find the main drivers of a problem, but standardization more often of the norm.
